# 3. Ampliando el dataframe
Al final de este proceso ampliaremos el df con 6 columnas nuevas que nos darán información sobre el contenido de cada juego

## Despliegue
El siguiente paso en el tratamiento del dataframe es el procesado de la columna 'tags'. Dado que gran parte del objetivo del EDA es explorar los géneros de los juegos, será necesario "desplegar" el diccionario que tienen en dicha columna.<br>
Para ello, crearemos nuevas columnas con el nombre de las etiquetas y las rellenaremos con la cantidad de votos que ha recibido cada una.

```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv('data/processed/MainDF_processed.csv').drop(labels='Unnamed: 0', axis=1)
df
```

```python
# Para poder trabajar con los diccionarios será transformalos a JSON, y para que el JSON los identifique correctamente es necesario que las claves string estén entre comillas dobles
df.tags = df.tags.str.replace("'",'"')
```

```python
# Instanciamos un conjunto vacío donde iremos añadiendo las distintas etiquetas
tag_names = set()
```

```python
# Recorremos toda la columna tags, nos quedamos con sus claves y las guardamos en el set tag_names
for i in df.tags:
    try:
        a = json.loads(i).keys()
        for j in a:
            try:
                tag_names.add(j)
            except: pass
    except: pass
```

```python
# Creo las nuevas columnas. Por defecto, indico que tendrán 0 votos
for i in tag_names:
    df[i] = 0
```

```python
# ¿Cuántas nuevas columnas he creado?
len(tag_names)
```

```python
# Ahora asigno a cada nueva columna los votos de sus etiquetas
for i in df.index:
    try:
        a = json.loads(df.loc[i,'tags'])
        for j in a:
            df.loc[i,j] = a[j]
    except: pass
```

```python
df
```

```python
# Guardamos el dataframe
df.to_csv('data/processed/MainDF_maxi.csv', index=False)
```

## Repliegue
Trabajar con semejante cantidad de columnas es muy engorroso, así que vamos a hacer un resumen de acuerdo a 6 de las categorías de la página Steamworks: https://partner.steamgames.com/doc/store/tags<br>
Nos quedaremos con la etiqueta más votada dentro de cada categoría.

```python
# Creamos las categorías y las guardamos en constantes.py
# Se muestran aquí para informar al lector
top_level_genres = ['Action','Adventure','Casual','Experimental','Puzzle','Racing','RPG','Simulation','Sports','Strategy','Tabletop']
genres = ['Action RPG','Action-Adventure','Arcade','Auto Battler','Automobile Sim','Base Building','Baseball','Basketball','Battle Royale',
          'BMX','Board Game','Bowling','Building','Card Game','Character Action Game','Chess','Clicker','Cycling','Diplomacy','eSports','Experimental',
          'Exploration','Farming Sim','Fighting','Football','God Game','Golf','Hacking','Hidden Object','Hockey','Idler','Interactive Fiction'
          'Management','Match 3','Medical Sim','Mini Golf','Mining','MMORPG','MOBA','Motocross','Open World','Outbreak Sim','Party-Based RPG',
          'Pinball','Platformer','Point & Click','Rhythm','Roguelike','RTS','Sandbox','Shooter','Skateboarding','Skating','Skiing','Snowboarding',
          'Soccer','Space Sim','Stealth','Strategy RPG','Survival','Tennis','Tower Defense','Trivia','Turn-Based Strategy','Visual Novel',
          'Walking Simulator','Word Game','Wrestling']
subgenres = ['2D Fighter','2D Platformer','3D Fighter','3D Platformer','4X','Action Roguelike','Arena Shooter',"Beat 'em up",'Bullet Hell',
             'Card Battler','Choose Your Own Adventure','City Builder','Collectathon','Colony Sim','Combat Racing','CRPG','Dating Sim',
             'Dungeon Crawler','Education','Flight','FPS','Grand Strategy','Hack and Slash','Heist','Hero Shooter','Horror','Immersive Sim',
             'Investigation','JRPG','Life Sim','Looter Shooter','Metroidvania','Mystery Dungeon','On-Rails Shooter','Open World Survival Craft',
             'Political Sim','Precision Platformer','Programming','Real Time Tactics','Roguelite','Roguevania','Runner',"Shoot 'Em Up",'Side Scroller',
             'Sokoban','Solitaire','Souls-like','Spectacle fighter','Spelling','Survival Horror','Tactical RPG','Third-Person Shooter',
             'Time Management','Top-Down Shooter','Trading','Trading Card Game','Traditional Roguelike','Turn-Based Tactics','Twin Stick Shooter','Typing','Wargame']
visuals = ['2.5D','2D','360 Video','3D','3D Vision','Abstract','Anime','Cartoon','Cartoony','Cinematic','Colorful','Comic Book',
           'Cute','First-Person','FMV','Hand-drawn','Isometric','Minimalist','Noir','Pixel Graphics','Psychedelic','Realistic',
           'Split Screen','Stylized','Text-Based','Third Person','Top-Down','Voxel','VR']
players = ['4 Player Local','Asynchronous Multiplayer','Co-op','Co-op Campaign','Local Co-Op','Local Multiplayer','Massively Multiplayer',
           'Multiplayer'',Online Co-Op','Singleplayer']
assesments = ['Addictive','Beautiful','Classic','Competitive','Cult Classic','Difficult','Emotional','Epic','Fast-Paced','Funny','Great Soundtrack',
              'Lore-Rich','Masterpiece','Psychological','Relaxing','Replay Value','Short','Unforgiving','Comedy','Dark Comedy']
```

```python
# Creamos las nuevas columnas. Las instanciamos con nans
df['Top level genre'] = pd.NA
df['Genres'] = pd.NA
df['Subgenres'] = pd.NA
df['Visuals'] = pd.NA
df['Players'] = pd.NA
df['Assesments'] = pd.NA
```

```python
# Este es el rango donde se encuentran las columnas de las que obtendremos los votos
df.loc[:,'Mature':'Parody ']
# Usando .columns:
df.columns[8:-7]
```

```python
# Con esta línea de código obtenemos una serie con las etiquetas y votos distintos de 0 en el rango de columnas mencionado anteriormente
# Ejemplo para el primer juego:
df.loc[0,(df.columns[8:-7][df.loc[0,'Mature':'Parody '] != 0])]
```

Ya está todo listo para recoger los votos y determinar cuál es la etiqueta más votada de cada categoría

```python
# Filtramos la serie por las categorías instanciadas y nos quedamos con el índice del valor máximo
# CUIDADO CON EJECUTAR ESTA CELDA, TARDA 5'
for i in df.index:
    tags = df.loc[i,(df.columns[8:-7][df.loc[i,'Mature':'Parody '] != 0])]
    try:
        visual_tag = tags[tags.index.isin(visuals)].idxmax()
        df.loc[i, 'Visuals'] = visual_tag
    except: pass
    try:
        players_tag = tags[tags.index.isin(players)].idxmax()
        df.loc[i, 'Players'] = players_tag
    except: pass
    try:
        top_level_genres_tag = tags[tags.index.isin(top_level_genres)].idxmax() 
        df.loc[i, 'Top level genre'] = top_level_genres_tag
    except: pass
    try:
        assesments_tag = tags[tags.index.isin(assesments)].idxmax()
        df.loc[i, 'Assesments'] = assesments_tag
    except: pass
    try:
        genres_tag = tags[tags.index.isin(genres)].idxmax()
        df.loc[i, 'Genres'] = genres_tag
    except: pass
    try:
        subgenres_tag = tags[tags.index.isin(subgenres)].idxmax()
        df.loc[i, 'Subgenres'] = subgenres_tag
    except: pass
```

```python
# Echamos un vistazo a nuestro nuevo dataframe excluyendo las columnas de los votos
df_full = df.loc[:,['ID','Game','Release date','Price','Owners','Developer(s)','Publisher(s)','Top level genre','Genres','Subgenres','Assesments','Visuals','Players']]
df_full
```

```python
# Evidentemente, ahora tendremos nulos en las categorías porque no todos los juegos tienen etiquetas para cada categoría
df_full.info()
```

```python
# De hecho, la columna Assesments tiene tan poca información que directamente la eliminamos
df_full.drop('Assesments', axis=1, inplace=True)
```

```python
df_full.info()
```

De las dos nuevas columnas, las más importantes son 'Top level genre' y 'Genres'. Estos son los campos que con más probabilidad deberían tener un dato. Si no lo tienen, es muy probable que el juego no sea popular en absoluto, y por lo tanto no es relevante en la exploración.<br>
Procedemos a eliminar aquellas filas que tengas estas dos columnas con un NaN

```python
df_full.dropna(axis=0, how='all', subset=['Top level genre', 'Genres'], inplace=True, ignore_index=True)
df_full.info()
```

```python
# Guardamos el dataframe en un nuevo CSV
df_full.to_csv('data/processed/MainDF_full.csv', index=False)
```
